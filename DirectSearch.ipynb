# Direct Search Method
In this  notebook, we discuss direct search method

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))

x = np.linspace(-20,20)
y = 2*x**2 + 20*x - 22
plt.plot(x,y)
plt.title(f'The plot of $2x^2 + 20x - 22$')
plt.xlabel('X-axis', fontsize = 18)
plt.ylabel('Y-axis', fontsize = 18)
plt.grid()
plt.show()
```

```python
def f(x):
    return 2*x**2 + 20*x - 22
```

```python
def direct_method(a,b,delta,e):
    """[summary]
    This function compute min of f(x)
    using direct method
    Args:
        a ([type]): [description]
        b ([type]): [description]
        delta ([type]): [description]
        e ([type]): [description]
    """
    step = 1
    print(f'\n\n*** Direct Search Method***')
    condition= True 
    while condition:
        x1 = (a+b-delta)/2
        x2 = (a+b+delta)/2
        print('Iteration-%d, x1 = %0.6f, x2 = %0.6f,'
              'f(x1) = %0.8f and f(x2) = %0.8f' % (step, x1, x2, f(x1),f(x2)))
        
        if f(x1) < f(x2):
            b = x2 
        elif f(x1) > f(x2):
            a = x1 
        else:
            a, b = x1, x2
            
        print(f'a = {a} and b = {b}')
    
           
        step += 1
        condition = step < 100
    print(f'\n Required Required Root is : %0.8f {(a+b)/2}') 
```

```python
direct_method(-7,-1,0.5,0.01)
```

```python
direct_method(-7,3,0.5,0.01)
```
