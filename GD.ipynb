# Gradient Descent

We will implement a simple form of Gradient Descent. We do it for the funtion $f(x) = x^3-4x^2+6.$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function

f = lambda x: x**3 - 4*x**2 + 6
x = np.linspace(-1,4,1000)
plt.plot(x,f(x), c='r')
plt.title('Cubic curve')
plt.show()
```

```python
f_der = lambda x: 3*x**2 - 8*x 
```

```python
def plot_grad(x,y,x_vis, y_vis):
    plt.subplot(1,2,2)
    plt.scatter(x_vis,y_vis, c='b')
    plt.plot(x,f(x), c = 'r')
    plt.title('Gradient Descent')
    plt.show()
    plt.subplot(1,2,1)
    plt.scatter(x_vis,y_vis, c="b")
    plt.plot(x,f(x), c='r')
    plt.xlim([2.5,3.0])
    plt.title('Zoomed in Figure')
    plt.show()

def grad_itr(x_start, itr, lrate):
    x_grad = [x_start]
    y_grad = [f(x_start)]
    
    # Looping until number of iterations
    for i in range(itr):
        x_start_grad = -f_der(x_start)
        # next point
        x_start += lrate*x_start_grad
        
        x_grad.append(x_start)
        y_grad.append(f(x_start))
    print("Local minimum occurs at : {:.2f}".format(x_start))
    print(f'Number of steps: {len(x_grad)-1}')
    plot_grad(x,f(x), x_grad, y_grad)
```

```python
grad_itr(0.5, 1000, 0.05)
```
